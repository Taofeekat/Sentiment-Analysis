```python
from textblob import TextBlob
import pandas as pd

# Load the dataset
data = pd.read_csv('DataDNA Dataset Challenge - March 2023.csv')

# printing the first 5 rows of the file
data.head(5)
```

```python
# checking the data columns row
data.info()
```

```python
#deleting the rows with null values 
data.dropna(inplace=True)
```

Method one

```python
# Define a function to perform sentiment analysis on each column
def get_sentiment_column(column):
    # Perform sentiment analysis on each value in the column
    sentiments = column.apply(lambda x: TextBlob(x).sentiment.polarity)  
    return sentiments

# Apply the sentiment analysis function to each column
data['sentiment_review_title'] = get_sentiment_column(data['review_title'])
data['sentiment_review_text'] = get_sentiment_column(data['review_text'])

# Save the result to a new CSV file and save on jupyter files
data.to_csv('datadna_dataset_with_sentiment.csv', index=False)
```

```python
# reading and previewing the newly saved sentiment data
new_data = pd.read_csv('datadna_dataset_with_sentiment.csv')
new_data.head(10)
```

Method 2

```python
# importing packages for sentimental analysis
from textblob import TextBlob
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Preprocess the data
stopwords = nltk.corpus.stopwords.words('english')
data['cleaned_text'] = data['review_text'].apply(lambda x: ' '.join([word.lower() for word in x.split() if word.lower() not in stopwords]))
data['cleaned_title'] = data['review_title'].apply(lambda x: ' '.join([word.lower() for word in x.split() if word.lower() not in stopwords]))

# Perform sentiment analysis
data['text_sentiment'] = data['cleaned_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
data['title_sentiment'] = data['cleaned_title'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Print the first 5 rows with sentiment score
print(data[['review_text', 'text_sentiment', 'title_sentiment']].head(5))
```

After examining the outcomes of both approaches, it can be seen that there are minor differences.
